## Plot

```python
import numpy as np
import matplotlib.pyplot as plt
```

```python
fig, axs = plt.subplots(1,2, figsize = (10, 10))
models = ['Ridge Regression', 'Resnet Pretrained', 'Resnet Randomly initialized', 'CNN', 'model 1', 'model 2', 'model 3']
evs_mean = [1, 2, 3, 4, 5, 6, 7]
corr_mean = [1, 2, 3, 4, 5, 6, 7]


bar_labels = ['data driven', 'task driven', 'data driven', 'task driven', 'data driven', 'task driven', 'task driven']
bar_colors = ['tab:red', 'tab:blue', 'tab:red', 'tab:blue', 'tab:blue', 'tab:red', 'tab:blue']

axs[0].bar(models, evs_mean, label=bar_labels, color=bar_colors)
axs[0].set_ylabel('Mean explain variance')
axs[0].set_title('Mean explain variance')
axs[0].legend(title='approachs')

axs[1].bar(models, corr_mean, label=bar_labels, color=bar_colors)
axs[1].set_ylabel('Mean correlation score')
axs[1].set_title('Mean correlation score')

plt.show()
```
